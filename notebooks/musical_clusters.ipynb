```python
full_df = pd.read_csv('../data/spot_lyr_df.csv')[['name_track', 'artist', 'name_album', 'clean_title', 'danceability', 'energy', 'key',\
                                                  'loudness', 'mode', 'speechiness',\
                                                  'acousticness', 'instrumentalness',\
                                                  'liveness', 'valence', 'tempo','time_signature']]
```

```python
full_df[full_df.artist == 'Eminem'].name_album.unique()
```

```python
unique_tracks = full_df.drop_duplicates(subset=['clean_title', 'artist'])
```

```python
unique_tracks
```

```python
unique_tracks[['danceability', 'energy', 'key','loudness', 'mode',\
               'speechiness','acousticness', 'instrumentalness',\
                'liveness', 'valence', 'tempo','time_signature']]
```

```python
from sklearn.preprocessing import StandardScaler
features = ['danceability', 'energy', 'key','loudness', 'mode',\
            'speechiness','acousticness', 'instrumentalness',\
            'liveness', 'valence', 'tempo','time_signature']
# Separating out the features
x = unique_tracks.loc[:, features].values

# Standardizing the features
x = StandardScaler().fit_transform(x)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
tracks_pca = pca.fit_transform(x)
pca_df = pd.DataFrame(data =list(zip(tracks_pca[:,0],tracks_pca[:,1])), columns=['C1', 'C2'])
pca_df['artist'] = list(unique_tracks['artist'])
pca_df['pca'] = pca_df.apply(lambda row:(row['C1'],row['C2']), axis=1)
```
