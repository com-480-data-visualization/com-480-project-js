```python
import pandas as pd
```

```python
ARTIST_NAME = 'Michael Jackson'
```

```python
lyrics = pd.read_csv('data/Lyrics_{0}_clean.csv'.format(ARTIST_NAME.replace(' ', '')))
```

```python
music = pd.read_pickle('data/{0}_tracks.pickle'.format(ARTIST_NAME.lower().replace(' ','_')))
```

```python
just_music = music[~(music.name_album == 'XSCAPE - Track by Track Commentary')]
just_music = just_music[~(just_music.name_track.map(lambda track: track.lower()).str.contains('mix'))]
just_music = just_music[~(just_music.name_track.map(lambda track: track.lower()).str.contains('interview'))]
just_music = just_music[~(just_music.name_track.map(lambda track: track.lower()).str.contains('excerpt'))]
```

```python
just_music
```

```python
clean_lyrics = lyrics[~(lyrics.title.map(lambda title: title.lower()).str.contains('mix'))]
```

```python
import re

def remove_superflous(text):
    text = re.sub(r'\([^)]*\)', '', text).split('-')[0]
    while text[0] == ' ':
        text = text[1:]
    while text[-1] == ' ':
        text = text[:-1]
        
    return text.lower().replace("'", "").replace("’", "").replace('  ', ' ').replace(' ', '_')
                                              
just_music['clean_title'] = just_music.name_track.map(lambda track: remove_superflous(track))
clean_lyrics['clean_title'] = clean_lyrics.title.map(lambda title: remove_superflous(title))
```

```python
missing_titles= (set(just_music.clean_title.unique())) - (set(just_music.clean_title.unique()) & set(clean_lyrics.clean_title.unique()))
missing_titles
```

```python
is_it_scary_lyrics = open('processbook/missing_lyrics/is_it_scary.txt', 'r').read()
watching_me_lyrics = open('processbook/missing_lyrics/watching_me.txt', 'r').read()
torture_lyrics = open('processbook/missing_lyrics/torture.txt', 'r').read()
```

```python
clean_lyrics = clean_lyrics.drop_duplicates(subset='clean_title')[['clean_title', 'lyrics']]
```

```python
just_music = just_music.merge(clean_lyrics, how='left', on='clean_title')
just_music.head()
```

```python
just_music.loc[just_music.clean_title.isin(missing_titles), 'lyrics'] = [watching_me_lyrics, torture_lyrics, is_it_scary_lyrics, is_it_scary_lyrics]
```

```python
just_music.to_csv('data/{0}_full_df.csv'.format(ARTIST_NAME.lower().replace(' ', '_')))
```
